# ATM Function

```python
Net_balance = 0.0
def atm():
    from datetime import datetime
    print(datetime.now(),"\nATM Service! \n\nWelcome",user_name)
    Opr = input("Please Type Any Option Provided Below And Press Enter. \n1. Check Account Balance \n2. Deposit \n3. Withdraw \n0. Exit \n") 
    while int(Opr) != 0:
        global Net_balance
        if int(Opr) == 1:
            print ("\nYour Acount Balance = Rs",Net_balance,"\n")
            
        elif int(Opr) == 2:
            Deposit()
            
        elif int(Opr) == 3:
            Withdraw()
            
        else:
            print ("Wrong Option!")
        
        Opr = input("1. Check Account Balance \n2. Deposit \n3. Withdraw \n0. Exit \n")
    print ("Thanks For Using ATM! \nWe Hope You Are Satisfied With Our Service.\nHave A Nice Day Ahead.")
    return
    
    
def Deposit():
    deposit = input("Enter Amount In Rupees ")
    if float(deposit) >= 0:
        global Net_balance
        Net_balance += float(deposit)
        return
    else:
        print ("\nPlease Enter Right Amount! \n")
        return Deposit()
        
def Withdraw():
    global Net_balance
    if float(Net_balance) <= 0:
        print ("\nWithdrawl Impossible! \nYour Account Balance = Rs",Net_balance,"\nPlease Deposit Amount First!")
        return
    else:
        withdraw = input("Enter Amount In Rupees ")
        if float(withdraw) <= float(Net_balance):
            if float(withdraw) > 0:
                Net_balance -= float(withdraw)
                return
            else:
                print ("\nPlease Enter Right Amount! \n")
                return Withdraw()
                    
        else:
            print ("\nWithdrawl Impossible! \nYour Acount Balance = Rs",Net_balance,"\n")
            return Withdraw()
```

# Login To ATM

```python
user_name = ""
filename = ""
d = {}

import os

def login_user(d):
    data()
    user = input("Please Select One \n1. Login \n2. Create New Account \n0. Exit \n")
    
    if int(user) == 1:
        login()
    
    elif int(user) == 2:
        new_account()
    
    elif int(user) == 0:
        print ("Good Bye!")
    
    else:
        return login_user()

def data():
    global filename
    global d
    directory = "/home/faizanf33/Desktop/ATM"
    name = "usersdata.txt"
    filename = os.path.join(directory, name)
    with open(filename, "r") as f:
        l = f.read().split(',')
        for i in l:
            a = i.split(":")
            d[a[0]] = a[1]
    return d
    
    
    
def login():
    global user_name
    user_name = input("Name : ")
    entry = 0
    if user_name in d.keys():
        while int(entry) != 3:
            pin = str(input("Enter 4-Digit Pin : "))
            if pin == d[user_name]:
                return atm()
            else:
                entry += 1
                print ("Incorrect Pin")
        print ("Login Unsuccessful")
        return login_user(d)        
    else:
        print ("Invalid User")
        return login_user(d)

    
def new_account():
    print (filename)
    user_name = input("Please Type Your Name : ")
    pin_count = 0
    while pin_count != 3:
        pin = str(input ("Enter 4-Digit Pin : "))
        confirm_pin = str(input ("Confirm Pin : "))
        if pin == confirm_pin:
            print ("Account Name :",user_name)
            confirm = input("Please Confirm \n1. Yes \n2. No \n")
                
            if int(confirm) == 1:  
                with open(filename, "a") as f:
                    w = ","+user_name+":"+pin
                    f.write(w)
                    f.close()
                    print ("Account Created Successfully! \n")
                    return login_user(d)
            elif int(confirm) == 2:
                return new_account()
            
        else:
            print ("Your Pin Did Not Match!")
            pin_count +=1
    print ("Account Not Created!")
    return login()
```
